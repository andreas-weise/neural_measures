```python
# load fake session and correlation test results
with open(cfg.SES_TEST_RES_FNAME, 'rb') as pickle_file:
    ses_res = pickle.load(pickle_file)
with open(cfg.COR_TEST_RES_FNAME, 'rb') as pickle_file:
    cor_res = pickle.load(pickle_file)
```

```python
# fake session test results
for m in cfg.MEASURE_IDS:
    for corpus_id in cfg.CORPUS_IDS:
        print('%s %s: %.3f (%.3f)' % 
              (m, corpus_id, 
               np.mean(ses_res[corpus_id][m]),
               np.std(ses_res[corpus_id][m])))
```

```python
# the same networks are used for the 100 runs of both tests to check
# between the results

# fake session recognition rates for both corpora correlate
x = ses_res[cfg.CORPUS_ID_FISHER][cfg.MEASURE_ID_DR]
y = ses_res[cfg.CORPUS_ID_GAMES][cfg.MEASURE_ID_DR]
scipy.stats.pearsonr(x,y)
```

```python
# performance on the fake session detection test does not correlate
# with social variable correlations, neither for signed correlations...
for m in cfg.MEASURE_IDS:
    for corpus_id in cfg.CORPUS_IDS:
        for i, v in enumerate(cfg.SOCIAL_VARS):
            x = [res[m][i][1] for res in cor_res]
            y = ses_res[corpus_id][m]
            print(m, corpus_id, v[0], scipy.stats.pearsonr(x,y))
```

```python
# ... nor for absolute values of the correlations
# (some correlations are always positive so abs is the same as above)
for m in cfg.MEASURE_IDS:
    for corpus_id in cfg.CORPUS_IDS:
        for i, v in enumerate(cfg.SOCIAL_VARS):
            x = [abs(res[m][i][1]) for res in cor_res]
            y = ses_res[corpus_id][m]
            print(m, corpus_id, v[0], scipy.stats.pearsonr(x,y))
```
